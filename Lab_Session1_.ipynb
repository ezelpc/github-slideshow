```python
import numpy as np

#  Se genera la lista de 100 numeros flotantes aleatorios entre -1000 y 1000
num_alt= np.random.uniform(-1000, 1000, 100)

# Imprime la lista de numeros que se generaron
print(num_alt)

# encontramos los numeros mas pequeños
dos_p = np.partition(num_alt, 2)[:2]

# encongramos los numeros mas grandes
dos_g = np.partition(num_alt, -2)[-2:]

print("Dos números más pequeños:", dos_p)
print("Dos números más grandes:", dos_g)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Se define la función f(t)
def f(t):
    return np.sin(2 * np.pi * t / 5 + 0.2)

# Se establece un rango para los dos periodos
t = np.linspace(0, 20, 1000)  # ajustar el rango dependiendo la necesidad

# Calcula los valores de f(t)
f_v= f(t)

# Encontrar los puntos donde la tangente tiene la pendiente más cercana a cero
secante = np.diff(f_v) / np.diff(t)
closest_to_zero_indices = np.where(np.abs(secante) == np.min(np.abs(secante)))

# Graficar f(t)
plt.figure(figsize=(10, 6))
plt.plot(t, f_v, label='f(t) = sin(2πt/5 + 0.2)')
plt.scatter(t[closest_to_zero_indices], f_v[closest_to_zero_indices], c='orange', label='Tangente más cercana a cero', marker='o')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.grid()
plt.title('Gráfico de f(t) y puntos donde la tangente tiene pendiente cercana a cero')
plt.show()
```
